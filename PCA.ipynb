```
import numpy as np
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.metrics import mean_squared_error as ms
```

```
def read_data (datafile):
  return np.loadtxt(datafile, skiprows = 1, delimiter = ',')
```

```
def train(training_file, para_file, n_PCs):  
  # Number of PC
  pca = PCA(n_components = n_PCs)
  # Train PCA model
  model = pca.fit(para_file)
  # Apply transformatio to target dataset
  data_trained = model.transform(training_file)
  # Project back to original size
  inverse_back = model.inverse_transform(data_trained)
  return inverse_back
```

```
def evaluate(transformed_data, true_data):
  return ms(transformed_data, true_data,multioutput = 'raw_values').sum()
```

```
def main():
  data1 = read_data ('drive/My Drive/CS498/HW3/dataI.csv')
  data2 = read_data ('drive/My Drive/CS498/HW3/dataII.csv')
  data3 = read_data ('drive/My Drive/CS498/HW3/dataIII.csv')
  data4 = read_data ('drive/My Drive/CS498/HW3/dataIV.csv')
  data5 = read_data ('drive/My Drive/CS498/HW3/dataV.csv')
  dataIris = read_data ('drive/My Drive/CS498/HW3/iris.csv')
  
  for i in range(5):
    count = 1
    for datafile in [data1, data2, data3, data4, data5]:
      data_transformed = train(datafile, dataIris, i)
      print('the result of Dataset',count, i, 'N:', evaluate(data_transformed, dataIris))
      count = count + 1
    print('\n')
    
  for i in range(5):
    count = 1
    for datafile in [data1, data2, data3, data4, data5]:
      data_transformed = train(datafile, datafile, i)
      print('the result of Dataset',count, i, 'c:', evaluate(data_transformed, dataIris))
      count = count + 1
    print('\n')    
```

```
main()
```
